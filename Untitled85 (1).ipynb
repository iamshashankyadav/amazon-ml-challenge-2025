```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    tokens = word_tokenize(str(text).lower())  # tokenize and lowercase
    filtered = [w for w in tokens if w not in stop_words]
    return ' '.join(filtered)  # join back into a string
df['Bullet Point 1'] = df['Bullet Point 1'].apply(remove_stopwords)
```

```python
def remove_non_ascii(text):
    """Encodes the text to ASCII, ignoring non-ASCII characters, then decodes back."""
    if pd.isna(text) or text == '':
        return ''
    # Encode to ASCII, using 'ignore' to drop non-ASCII chars (emojis, accents, etc.)
    # Then decode back to a standard string
    return text.encode('ascii', 'ignore').decode('ascii')

# Apply the ASCII removal function to the consolidated column
df['Bullet Point 1'] = df['Bullet Point 1'].apply(remove_non_ascii)

# 3. Final cleanup of double spaces (sometimes left behind by filtering)
df['Bullet Point 1'] = df['Bullet Point 1'].str.replace(r'\s{2,}', ' ', regex=True).str.strip()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import re

try:
    df = pd.read_csv('output(42) (5).csv')

    # --- Preprocessing ---
    # Handle missing values in 'Bullet Point 1'
    df['Bullet Point 1'].fillna('', inplace=True)

    # Convert to lowercase
    df['Bullet Point 1'] = df['Bullet Point 1'].str.lower()

    # Remove punctuation
    df['Bullet Point 1'] = df['Bullet Point 1'].apply(lambda x: re.sub(r'[^\w\s]', '', x))

    # --- TF-IDF Vectorization ---
    # Initialize the TfidfVectorizer with English stop words
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')

    # Fit and transform the processed text data
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['Bullet Point 1'])

    # Get the feature names (the words)
    feature_names = tfidf_vectorizer.get_feature_names_out()

    # --- Extracting Important Words ---
    def get_important_words(tfidf_row, feature_names, top_n=5):
        """
        Get the top_n important words from a tfidf_row based on highest scores.
        """
        top_indices = tfidf_row.argsort()[-top_n:][::-1]
        important_words = [feature_names[i] for i in top_indices]
        return ' '.join(important_words)

    # Apply the function to each row of the tfidf_matrix
    df['important_words'] = [get_important_words(row.toarray()[0], feature_names) for row in tfidf_matrix]

    # --- Display Results ---
    print("Original Bullet Points and Extracted Important Words:")
    for index, row in df.iterrows():
        print(f"--- Entry {index+1} ---")
        print(f"Original: {row['Bullet Point 1']}")
        print(f"Important Words: {row['important_words']}")
        print("\n")

    # Save the results to a new CSV file
    output_df = df[['Bullet Point 1', 'important_words']]
    output_df.to_csv('important_words_output.csv', index=False)
    print("Results saved to 'important_words_output.csv'")

except FileNotFoundError:
    print("Error: 'first10 (1).csv' not found. Please make sure the file is in the same directory as your script or notebook.")
except Exception as e:
    print(f"An error occurred: {e}")
```
