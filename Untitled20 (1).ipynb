```python
import pandas as pd
import re

# Assuming df is your DataFrame
# and catalog_content column has lists like the one you showed
def parse_catalog(catalog_list):
    data = {}
    bullet_points = []

    for item in catalog_list:
        if item.startswith("Item Name:"):
            data["Item Name"] = item.split("Item Name:", 1)[1].strip()

        elif re.match(r"Bullet Point\s*\d*:?", item):
            # Extract text after the colon
            bullet_points.append(item.split(":", 1)[1].strip())

        elif item.startswith("Product Description:"):
            data["Product Description"] = item.split("Product Description:", 1)[1].strip()

        elif item.startswith("Value:"):
            data["Value"] = item.split("Value:", 1)[1].strip()

        elif item.startswith("Unit:"):
            data["Unit"] = item.split("Unit:", 1)[1].strip()

    # Join bullet points as one string (or keep separate if you prefer)
    for idx, bullet in enumerate(bullet_points, start=1):
        data[f"Bullet Point {idx}"] = bullet

    return data

# Apply to each row in DataFrame
parsed = df["catalog_cleaned"].apply(parse_catalog)

# Convert list of dicts to DataFrame
parsed_df = pd.DataFrame(parsed.tolist())

# Merge back with main DataFrame
final_df = pd.concat([df.drop(columns=["catalog_cleaned", "catalog_content"]), parsed_df], axis=1)
```
